```python
import pandas as pd
import numpy as np
```

```python
drug = pd.read_csv("/content/drive/MyDrive/2021캡스톤/drug_gene_preprocessed.csv")
drug
```

```python
drug = drug.drop(['index'], axis=1)
display(drug)
```

```python
drug_r = drug.drop(['measure_of_response', 'regimen_indication',	'therapy_types',	'drug_name_',	'therapy_response'], axis=1)
drug_r
```

```python
drug_r.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
drugs = drug_r[['bcr_patient_barcode', '?|100130426',	'?|100133144',	'?|100134869',	'?|10357',	'?|10431',	'?|136542' ]]
drugs
```

```python
drug_ri = drug_r
drug_ri.index = drug_r['bcr_patient_barcode']
drug_ri = drug_ri.drop(['bcr_patient_barcode'],axis=1)
drug_ri
```

```python
std = StandardScaler().fit_transform(drug_ri)
std = pd.DataFrame(std, index=drug_ri.index, columns=drug_ri.columns)
drug_ri = std
drug_ri.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca_array = pca.fit_transform(drug_ri)
pca_df = pd.DataFrame(data=pca_array, columns = ['pc1', 'pc2', 'pc3', 'pc4', 'pc5', 'pc6', 'pc7', 'pc8', 
                                                 'pc9', 'pc10'])
pca_df.head()
```

```python
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```
