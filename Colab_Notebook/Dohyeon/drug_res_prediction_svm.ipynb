```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
```

```python
df_svm = pd.read_csv("/content/drive/MyDrive/SW21/drug_gene_preprocessed.csv")
df_svm
```

```python
# ?로 시작하는거 없애줌
df_svm.drop(df_svm.columns[7:36], axis = 1, inplace = True)
df_svm.head(2)
```

```python
df_svm.drop(["index", "measure_of_response"], axis = 1, inplace = True)
df_svm = pd.get_dummies(data = df_svm, columns = ["regimen_indication", "therapy_types", "drug_name_"])
df_svm.head(2)
```

#### data target split

```python
data = np.array(df_svm[df_svm.columns.difference(["bcr_patient_barcode", "therapy_response"])])
target = np.array(df_svm.therapy_response)
```

```python
np.asarray(data).astype('float32')
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(target)
labels = encoder.transform(target)

target = labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target)
X_train.shape
```

```python
linear_svm = LinearSVC().fit(X_train, y_train)
```

```python
print("Coefficient : ", linear_svm.coef_)
print("Intercept : ", linear_svm.intercept_)
```

L1-based feature selection<br>
https://scikit-learn.org/stable/modules/feature_selection.html



```python
from sklearn.feature_selection import SelectFromModel

model = SelectFromModel(linear_svm, prefit = True)
X_new = model.transform(X_train)
X_new.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)

print("Accuracy on training set : {:.3f}".format(
    svc.score(X_train_scaled, y_train)))
print("Accuracy on test set : {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(df.drug_name_)
labels = encoder.transform(df.drug_name_)

df.drug_name_ = labels
df.head(10)
```

Label Encoding 결과<br>

* bevacizumab	: 0
* capecitabine : 1
* fluorouracil : 2
* folinic : 3
* irinotecan : 4
* oxaliplatin : 5

```python
data = df[df.columns.difference(["bcr_patient_barcode", "therapy_response"])]
target = df.therapy_response
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
linear_svm = LinearSVC().fit(X_train_scaled, y_train)
```

```python
print("Coefficient : ", linear_svm.coef_)
print("Intercept : ", linear_svm.intercept_)
```

```python
from sklearn.feature_selection import SelectFromModel

model = SelectFromModel(linear_svm, prefit = True)
X_train_new = model.transform(X_train_scaled)
X_train_new.shape
```

```python
linear_svm2 = LinearSVC().fit(X_train_new, y_train)

print("Coefficient : ", linear_svm2.coef_)
print("Intercept : ", linear_svm2.intercept_)
```

```python
model = SelectFromModel(linear_svm, prefit = True)
X_test_new = model.transform(X_test_scaled)
X_test_new.shape
```

```python
print("Accuracy on training set : {:.3f}".format(linear_svm2.score(X_train_new, y_train)))
print("Accuracy on test set : {:.3f}".format(linear_svm2.score(X_test_new, y_test)))
```

#### SVC()
